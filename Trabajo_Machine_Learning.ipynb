```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/1SJulioS1/Machine_learning/main/Datasets/Datasets/Breast%20cancer%20Wisconsin/breast-cancer.csv")
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df.drop('id', inplace=True, axis=1)
labelencoder = LabelEncoder()
df['diagnosis'] = labelencoder.fit_transform(df['diagnosis'])
df
```

```python
from sklearn.model_selection import train_test_split
X = df.loc[:, df.columns != "diagnosis"]
y = df.loc[:,'diagnosis']

X_train, X_test, y_train, y_test = train_test_split(
    X,  # columns for x
    y,  # column for y
    test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

X_train = std_scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train)
X_test = std_scaler.fit_transform(X_test)
X_test = pd.DataFrame(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


pipe_svc = make_pipeline(StandardScaler(),
                         SVC(random_state=1))

param_range = [0.1, 1.0, 10.0, 100.0, 1000.0]

grid = [{'svc__C': param_range, 'svc__gamma': param_range, 'svc__kernel': ['rbf']}]

gs = GridSearchCV(estimator=pipe_svc, param_grid=grid, scoring='accuracy', cv=10)
gs = gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
clf = gs.best_estimator_
print(f'Test accuracy: {clf.score(X_test, y_test):.3f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
grid = {'max_depth': [50,75,100,125,150,200,250]}

rf = RandomForestClassifier()
grid_forest =  GridSearchCV(estimator=rf, param_grid=grid, scoring='accuracy', cv=10)
gs = grid_forest.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
from sklearn import tree

dec_tree = tree.DecisionTreeClassifier()
grid = {"criterion":['gini','entropy'],"max_depth": range(1,30)}

gs = GridSearchCV(dec_tree, param_grid=grid, cv = 5, n_jobs=-1)

gs = gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
from sklearn.neural_network import MLPClassifier

nn=MLPClassifier(max_iter=100000000, alpha=0.0001, activation='tanh', hidden_layer_sizes=(10,17,17,17,17))
nn.fit(X_train, y_train)
```

```python
nn.score(X_test,y_test)
```
